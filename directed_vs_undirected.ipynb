```python
import numpy as np
import matplotlib.pyplot as plt

N = 10
all_e, all_H, all_2H = [],[], []
for e in range(N**2-N):
    all_e.append(e)
    e_2 = (N**2 if e*2>N**2 else e*2)
    H = - ((e/(N**2)) * np.log((e/(N**2))) + ((1-(e/(N**2))) * np.log(((1-(e/(N**2)))))))
    H_2 = - ((e_2/(N**2)) * np.log((e_2/(N**2))) + ((1-(e_2/(N**2))) * np.log(((1-(e_2/(N**2)))))))
    all_H.append(H)
    all_2H.append(H_2)
plt.figure(figsize=(10,7))
plt.scatter(all_e,all_H,c='g')
plt.scatter(all_e,all_2H,c='#63b175')
plt.xlabel('e')
plt.ylabel('H')
plt.legend(['H(d)','H(u)'])
plt.show()
```
